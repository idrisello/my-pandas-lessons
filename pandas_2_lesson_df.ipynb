```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
data={'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       "Temp":[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]}
df=DataFrame(data)
df
```

```python
df.head() #jadvalning boshidan 5 ta qator ko'rishimiz mumkin.
```

```python
df = DataFrame(data, columns=['Yil','Temp','Aholi soni']) #taFrame yaratishda har bir qator indekslanadi (0,1,2...), ustunlar ketma-ketligi esa lug'atdagi ketma-ketlikka bog'liq bo'ladi. Ustunlar ketma-ketligini berishni istasak DataFrame yaratishda columns parametrini ham ko'rsatib ketishimiz mumkin:
df
```

```python
#Agar columns ro'yxarida lug'atda mavjud bo'lmagan ustin bo'lsa, shu ustunga bo'sh qiymatlar qo'yib ketiladi.
df = DataFrame(data, columns=['Yil','Temp','Aholi soni','YAIM'])
df.head()
```

```python
#Series kabi DataFrame indekslariga ham o'zimiz qiymat berishimiz mumkin:
df = DataFrame(data, columns=['Yil','Temp','Aholi soni','YAIM'], index=[21, 20, 19, 18, 17, 16, 15, 10])
df
```
