```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 't'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma','Anor','Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('abc'), index=['Olma','Anor','Qovun','Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1+df2 #DataFrameda qator va ustun indekslari hisobga olinadi.
```

```python
df1.add(df2,fill_value=0) #ll_value parametri yordamida mos tushuvchi indekslardagi NaN qiymatlarni boshqa qiymat bilan almashtirish mumkin. Bu usulda yakuniy natijada ikkala df da ham mavjud bo'lmagan indekslarda NaN qiymati qoladi.
```

# Arifmetik amallar


*   add, radd - Qo'shish (+)


*   sub, rsub - Ayirish (-)
*   div, rdiv - Bo'lish (/)



*   floordiv, rfloordiv - Bo'lish va butun qismini olis (//)
*   pow, rpow - Eksponenta (**)




*   mul, rmul - Ko'paytirish (*)

```python
df = pd.DataFrame(np.arange(1,13.).reshape(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
df
```

```python
obj = pd.Series([1,1,2,2], index=df.columns)
obj
```

```python
df - obj #series indekslari datafranega indekslariga nisbatan ayriladi
```

```python
df - obj['Olma'] #df barcha qiymatlaridan olmaning qiymati ayriladi
```

```python
# DF qatori max va min qiymati ayirmasini hisoblaydigan funksiya
f = lambda x: x.max() - x.min()
```

```python
def foo(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])
```
